```python
url = "https://hubeau.eaufrance.fr/api/v2/hydrometrie/obs_elab"
params = {"code_entite": "A158020101", "size": 20}      
response_body = request_json(url, params=params)  # ignore page limitation
```

```python
print(type(response_body), response_body.keys())
assert(all(key in response_body.keys() for key in ("api_version", "count", "first", "prev", "next", "data")))
data = response_body["data"]
print(type(data), len(data))
assert(len(data) > 0)
data_0 = data[0]
print(type(data_0), data_0.keys())
```
