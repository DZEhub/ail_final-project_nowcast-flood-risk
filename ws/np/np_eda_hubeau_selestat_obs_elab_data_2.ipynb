# 0. Summary
- Download the file "hubeau_{HUBEAU_DATA_ENDPOINT}_{SITE}_{min_date}_{max_date}_1.csv" from LOCAL_DATA_DIRECTORY
- Convert "date_obs_elab" to datetime
- Pivot "grandeur" quantities to columns with values "resultat"
- Save to LOCAL_DATA_DIRECTORY as "hubeau_{HUBEAU_DATA_ENDPOINT}_{SITE}_{min_date}_{max_date}_2.csv" 

```python
import os
import pandas
import datetime

# LOCAL COMPONENTS:
from config import LOCAL_DATA_DIRECTORY, SITES
from utils import request_json_all, save_dataframe_as_csv
```

```python
print("Local data directory:", LOCAL_DATA_DIRECTORY)
assert(os.path.exists(LOCAL_DATA_DIRECTORY))
```

```python
HUBEAU_DATA_ENDPOINT = "obs_elab"
```

```python
SITE = SITES["Selestat"]["code"]
STATIONS = ",".join([station["code"] for station in SITES["Selestat"]["stations"]])
```

```python
print("Site:", SITE, "/", "Stations:", STATIONS)
```

# 1. Reload the data

```python
min_date, max_date = "20070101", "20260601"
basename = f"hubeau_{HUBEAU_DATA_ENDPOINT}_{SITE}_{min_date}_{max_date}"
```

```python
filepath_1 = os.path.join(LOCAL_DATA_DIRECTORY, f"{basename}_1.csv")
df = pandas.read_csv(filepath_1)
```

```python
df["date_obs_elab"] = pandas.to_datetime(df["date_obs_elab"])
df.sort_values(by="date_obs_elab", ascending=True, inplace=True)
display(df.head(20))
```

# 3. Pivot quantities

```python
df_2 = df.pivot_table(
    index=["code_site", "code_station", "date_obs_elab", "longitude", "latitude"],
    columns="grandeur_hydro_elab",
    values="resultat_obs_elab",
    aggfunc="first")
df_2 = df_2.reset_index()
df_2.columns.name = None
```

# 10. Save the dataframe

```python
filepath_2 = os.path.join(LOCAL_DATA_DIRECTORY, f"{basename}_2.csv")
df_2.to_csv(filepath_2, index=False, encoding="utf-8")
```
